# Define Preictal & Interictal Sections

## - Preictal

```python
# Crop the preictal section from chb01_15 epoch (Seizure onset in chb01_15: 1732 s, Preictal duration: 30 min = 1800 s)
seizure_onset = 3600 + 1732
preictal_duration = 1800
preictal_beginning = seizure_onset - preictal_duration

# Concatenating epochs chb01_15 and chb01_14
chb01_14_15 = mne.concatenate_raws([chb01_14_raw, chb01_15_raw])
preictal_section = chb01_14_15.copy().crop(tmin=preictal_beginning, tmax=seizure_onset, include_tmax=False)
```

```python
preictal_section_df = preictal_section.to_data_frame()

# Deleting last 5 unneeded channels
try:
    preictal_section_df = preictal_section_df.drop(columns=['P7-T7', 'T7-FT9', 'FT9-FT10', 'FT10-T8', 'T8-P8-1'])
except:
    print('There is no such channel in this epoch')
```

## - Interictal

```python
# Crop the interictal section from epochs chb01_10 ,chb01_11, chb01_12, chb01_13 and chb01_14 (Seizure onset in chb01_15 :1732 s, Interictal duration: 240 min = 14400 s, Excluded: 30 min = 1800 s)
seizure_onset = (5*3600) + 1732 
excluded = 2700
interictal_duration = 14400
interictal_beginning = seizure_onset - (preictal_duration + excluded + interictal_duration)
interictal_ending = interictal_beginning + interictal_duration

# Concatenating all 6 epochs
all_6_raws = mne.concatenate_raws([chb01_10_raw, chb01_11_raw, chb01_12_raw, chb01_13_raw, chb01_14_raw, chb01_15_raw])
interictal_section = all_6_raws.copy().crop(tmin=interictal_beginning, tmax=interictal_ending, include_tmax=False)
```

```python
interictal_section_df = interictal_section.to_data_frame()

# Deleting last 5 unneeded channels
try:
    interictal_section_df = interictal_section_df.drop(columns=['P7-T7', 'T7-FT9', 'FT9-FT10', 'FT10-T8', 'T8-P8-1'])
except:
    print('There is no such channel in this epoch')
```

# Segmentation

```python
# We have 1280 data points in each of our 5s segments
```

- ### Preictal

```python
segmented_preictal = []

for channel in list(preictal_section_df)[1:]:
    segmented_preictal.append(np.split(preictal_section_df[channel], 360))
```

- ### Interictal

```python
segmented_interictal = []

for channel in list(interictal_section_df)[1:]:
    segmented_interictal.append(np.split(interictal_section_df[channel], 2880))
```

# Segment Selection

```python
def random(high, size):
    rand = np.random.randint(0, high, size)
    while np.unique(rand).size != size:
        rand = np.random.randint(0, high, size)
    return rand
```

```python
# We want total number of 30 segments (15 preictal , 15 interictal)
```

```python
# Choosing 15 preictal segments out of 360 of them randomly
p_indices = random(360, 15)

# Choosing 15 interictal segments out of 2880 of them randomly
i_indices = random(2880, 15)
```

```python
segmented_preictal = np.array(segmented_preictal)
selected_preictal_segments = [segmented_preictal[:, i] for i in p_indices]
selected_preictal_segments = np.array(selected_preictal_segments)
```

```python
segmented_interictal = np.array(segmented_interictal)
selected_interictal_segments = [segmented_interictal[:, i] for i in i_indices]
selected_interictal_segments = np.array(selected_interictal_segments)
```

```python
np.array(selected_preictal_segments).shape, np.array(selected_interictal_segments).shape
```

# Feature Extraction & Classification

```python
# We have 2 features (mean, var)
# Let preictal and interictal be class 1 and 0 respectively
# We have 18 channel but the second dim is set 19 becuase the last one demonstrate segment class
```

- ### Preictal

```python
preictal_features = np.zeros((15, 19, 2))

for i in range(len(selected_preictal_segments)):
    for j in range(len(selected_preictal_segments[0])):
        preictal_features[i][j] = np.array([np.mean(selected_preictal_segments[i][j]), np.var(selected_preictal_segments[i][j])])
    preictal_features[i][j+1] = 1
```

```python
# Later we use np.unique on this column because we only need one of them
preictal_features[0][18]
```

- ### Interictal

```python
interictal_features = np.zeros((15, 19, 2))

for i in range(len(selected_interictal_segments)):
    for j in range(len(selected_interictal_segments[0])):
        interictal_features[i][j] = np.array([np.mean(selected_interictal_segments[i][j]), np.var(selected_interictal_segments[i][j])])
    interictal_features[i][j+1] = 0
```

```python
# Later we use np.unique on this column because we only need one of them
interictal_features[0][18]
```

# Merging All Of The Segments

```python
all_data = np.concatenate((preictal_features, interictal_features), axis=0)
np.random.shuffle(all_data)
```
